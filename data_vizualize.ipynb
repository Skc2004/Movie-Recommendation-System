```python
numerical_cols =df.select_dtypes(include=['number'])
numerical_cols=numerical_cols.drop(columns=['show_id','release_year','duration'])
mean_values = numerical_cols.mean()
median_values = numerical_cols.median()
mode_values = numerical_cols.mode().iloc[0]

print("Mean:\n", mean_values)
print("Median:\n",median_values)
print("Mode:\n", mode_values)
```

```python
min_values=numerical_cols.min()
max_values=numerical_cols.max()
range_values=max_values-min_values

print('Minimum Values:\n',min_values)
print('Maximum Values:\n',max_values)
print('Range Values:\n',range_values)
```

```python
standard_deviation=numerical_cols.std()
print('Standard Deviation:\n',standard_deviation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")

# --- 1. Histogram of Movie Ratings ---
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.show() # Close the plot to prevent it from displaying in a script environment

# --- 2. Bar Chart of Top 10 Countries ---
plt.figure(figsize=(12, 8))
# Clean data by removing rows where 'country' is missing
df_countries = df.dropna(subset=['country'])
country_counts = df_countries['country'].str.split(', ').explode().value_counts()
top_10_countries = country_counts.head(10)
sns.barplot(x=top_10_countries.values, y=top_10_countries.index, palette='viridis')
plt.title('Top 10 Countries by Number of Movies Produced')
plt.xlabel('Number of Movies')
plt.ylabel('Country')
plt.show()

# --- 3. Pie Chart of Movie Genres ---
plt.figure(figsize=(12, 12))
# Clean data by removing rows where 'genres' is missing
df_genres = df.dropna(subset=['genres'])
genre_counts = df_genres['genres'].str.split(', ').explode().value_counts()
# Group less frequent genres into 'Other' for clarity
top_genres = genre_counts.head(10)
other_count = genre_counts.iloc[10:].sum()
top_genres['Other'] = other_count
# Create the donut chart
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Proportion of Movie Genres')
plt.axis('equal')
plt.show()

# --- 4. Boxplot of Movie Popularity ---
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['popularity'], color='lightcoral')
plt.title('Distribution of Movie Popularity')
plt.xlabel('Popularity Score')
plt.xscale('log') # Use a log scale to better visualize the wide range of data
plt.show()

# --- 5. Line Chart of Movies Released Per Year ---
plt.figure(figsize=(12, 6))
movies_per_year = df['release_year'].value_counts().sort_index()
# Filter to show a relevant trend up to the last full year
movies_per_year = movies_per_year.loc[movies_per_year.index <= 2024]
sns.lineplot(x=movies_per_year.index, y=movies_per_year.values, marker='o', color='mediumseagreen')
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# --- 6. Scatter Plot of Rating vs. Popularity ---
plt.figure(figsize=(10, 6))
# Filter out movies with a rating of 0 for a more meaningful plot
df_scatter = df[df['rating'] > 0]
sns.scatterplot(x='vote_average', y='popularity', data=df_scatter, alpha=0.5, color='purple', edgecolor='w')
plt.title('Relationship Between Rating and Popularity')
plt.xlabel('Average Vote')
plt.ylabel('Popularity')
plt.yscale('log') # Use a log scale to handle the wide range of popularity values
plt.show()

```
